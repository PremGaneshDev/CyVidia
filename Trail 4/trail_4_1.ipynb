```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Download NLTK stopwords (you only need to do this once)
nltk.download('stopwords')

# Define a function for text cleaning
def clean_text(text):
    # Tokenize the text
    words = word_tokenize(text)
    
    # Remove punctuation and convert to lowercase
    words = [word.lower() for word in words if word.isalnum()]
    
    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    words = [word for word in words if word not in stop_words]
    
    # Join the words back into a clean text
    clean_text = ' '.join(words)
    
    return clean_text

# Apply the clean_text function to your Requirement Description column
df['Cleaned_Description'] = df['Requirement Description'].apply(clean_text)


# Configure pandas to display all rows and columns
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
# Configure pandas to display the entire column content
pd.set_option('display.max_colwidth', None)

# Print the entire DataFrame to check the results
print(df['Cleaned_Description'])
```

```python
# Import necessary libraries
import pandas as pd
from sklearn.model_selection import train_test_split

# Load your dataset
file_path = 'POC Based Requirements.xlsx'
df = pd.read_excel(file_path)

# Split the data into training (80%) and testing (20%) sets
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)

# Print unique labels in the training set
unique_train_buckets = train_df['Requirements Bucket'].unique()
unique_train_areas = train_df['Requirement Area'].unique()

print("Unique Requirement Buckets in Training Set:")
print(unique_train_buckets)

print("\nUnique Requirement Areas in Training Set:")
print(unique_train_areas)

# Print unique labels in the testing set
unique_test_buckets = test_df['Requirements Bucket'].unique()
unique_test_areas = test_df['Requirement Area'].unique()

print("\nUnique Requirement Buckets in Testing Set:")
print(unique_test_buckets)

print("\nUnique Requirement Areas in Testing Set:")
print(unique_test_areas)
```
